```python
# 生成日期
today = datetime.date.today()
year = datetime.datetime.today().year
month = datetime.datetime.today().month
day = datetime.datetime.today().day

today_str = f'{year}' + f'{month}'.zfill(2) + f'{day}'.zfill(2)
today_test = datetime.datetime.strptime(today_str, "%Y%m%d")

def date_diff(date_1, date_2):
    delta = datetime.datetime.strptime(date_2, "%Y%m%d") - datetime.datetime.strptime(date_1, "%Y%m%d")
    return delta.days
print(date_diff('20141231','20150103'))
```

```python
# 获取指数价格
index_ih = pro.index_daily(**{'ts_code':'000016.SH', "start_date": '20150416',},fields=["trade_date","open","close","high","low","change","amount"])
index_if = pro.index_daily(ts_code='000300.SH',fields=["trade_date","open","close","high","low","change","amount"])
index_ic = pro.index_daily(ts_code='000905.SH',fields=["trade_date","open","close","high","low","change","amount"])
index_ih.rename(columns = {'open':'open_ih','close':'close_ih','high':'high_ih','low':'low_ih', 'change':'change_ih', 'change':'change_ih', 'amount':'amount_ih'}, inplace = True)
index_if.rename(columns = {'open':'open_if','close':'close_if','high':'high_if','low':'low_if', 'change':'change_if', 'change':'change_if', 'amount':'amount_if'}, inplace = True)
index_ic.rename(columns = {'open':'open_ic','close':'close_ic','high':'high_ic','low':'low_ic', 'change':'change_ic', 'change':'change_ic', 'amount':'amount_ic'}, inplace = True)
data = pd.merge(index_ih, index_if, on='trade_date')
data = pd.merge(data, index_ic, on='trade_date')
data = data[::-1]
data
```

```python
# 获取股指期货信息
future_list = pro.fut_basic(exchange='CFFEX', fut_type='1', fields='ts_code,symbol,list_date,delist_date')
for i in range(len(future_list))[::-1]:
    if future_list.iat[i, 0][0] == 'T':
        future_list = future_list.drop(index=i)
def type(symbol):
    return symbol[0:2]
future_list['index_type'] = future_list['symbol'].map(type)

future_list.sort_values(by='symbol', ascending=True, inplace = True)

future_list
```

```python
# 生成合约代码
# 定义 近月合约 为maturity最短且大于5天的合约
# 定义 远月合约 为maturity最长的合约
min_maturity = 3

def current_contract(trade_date, index, type, future_list):
    future_list_1 = future_list[future_list['list_date']<=trade_date]
    future_list_2 = future_list_1[future_list_1['delist_date']>=trade_date]
    future_list_3 = future_list_2[future_list_2['index_type']==index]
    future_list_4 = future_list_3[future_list_3['index_type']==index]
    far = 0
    near = 300
    far_contract = ''
    near_contract = ''
    for i in range(len(future_list_4)):
        maturity = date_diff(trade_date,future_list_4['delist_date'].iloc[i])
        if maturity>=min_maturity:
            if maturity>far:
                far = maturity
                far_contract = future_list_4['ts_code'].iloc[i]
            if maturity<near:
                near = maturity
                near_contract = future_list_4['ts_code'].iloc[i]
    if type == 0:
        return near_contract
    return far_contract

data['near_contract_ih'] = data['trade_date'].map(lambda x: current_contract(x, 'IH', 0, future_list))
data['near_contract_if'] = data['trade_date'].map(lambda x: current_contract(x, 'IF', 0, future_list))
data['near_contract_ic'] = data['trade_date'].map(lambda x: current_contract(x, 'IC', 0, future_list))
data['far_contract_ih'] = data['trade_date'].map(lambda x: current_contract(x, 'IH', 1, future_list))
data['far_contract_if'] = data['trade_date'].map(lambda x: current_contract(x, 'IF', 1, future_list))
data['far_contract_ic'] = data['trade_date'].map(lambda x: current_contract(x, 'IC', 1, future_list))

```

```python
# 生成合约到期日
data = pd.merge(data, future_list[['ts_code','delist_date']], left_on='near_contract_ih',right_on='ts_code')
data.rename(columns = {'delist_date':'near_delist_date'},inplace=True)
data.drop(columns='ts_code',inplace = True)
data = pd.merge(data, future_list[['ts_code','delist_date']], left_on='far_contract_ih',right_on='ts_code')
data.rename(columns = {'delist_date':'far_delist_date'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data
```

```python
# 拉取IH数据
price_ih = pro.fut_daily(**{"trade_date": "","ts_code": 'IC1505.CFX'}, fields=["close","trade_date"])

for i in range(len(future_list)):
    name = future_list['ts_code'].iloc[i]
    if future_list['index_type'].iloc[i]=='IH':
        df = pro.fut_daily(**{"trade_date": "","ts_code": name}, fields=["trade_date","ts_code","close"])
        price_ih = pd.concat([price_ih, df], axis=0)
    
price_ih.dropna(inplace=True)
time.sleep(20)
price_ih
```

```python
# 拉取IF数据
price_if = pro.fut_daily(**{"trade_date": "","ts_code": 'IC1505.CFX'}, fields=["close","trade_date"])

for i in range(len(future_list)):
    name = future_list['ts_code'].iloc[i]
    if future_list['index_type'].iloc[i]=='IF':
        df = pro.fut_daily(**{"trade_date": "","ts_code": name}, fields=["trade_date","ts_code","close"])
        price_if = pd.concat([price_if, df], axis=0)
    
price_if.dropna(inplace=True)
time.sleep(20)
price_if
```

```python
# 拉取IC数据
price_ic = pro.fut_daily(**{"trade_date": "","ts_code": 'IC1505.CFX'}, fields=["close","trade_date"])

for i in range(len(future_list)):
    name = future_list['ts_code'].iloc[i]
    if future_list['index_type'].iloc[i]=='IC':
        df = pro.fut_daily(**{"trade_date": "","ts_code": name}, fields=["trade_date","ts_code","close"])
        price_ic = pd.concat([price_ic, df], axis=0)
    
price_ic.dropna(inplace=True)
price_ic
```

```python
#期货价格数据拼合

data = pd.merge(data, price_ih, left_on = ['trade_date', 'near_contract_ih'], right_on = ['trade_date', 'ts_code'])
data.rename(columns = {'close':'near_price_ih'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data = pd.merge(data, price_if, left_on = ['trade_date', 'near_contract_if'], right_on = ['trade_date', 'ts_code'])
data.rename(columns = {'close':'near_price_if'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data = pd.merge(data, price_ic, left_on = ['trade_date', 'near_contract_ic'], right_on = ['trade_date', 'ts_code'])
data.rename(columns = {'close':'near_price_ic'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data = pd.merge(data, price_ih, left_on = ['trade_date', 'far_contract_ih'], right_on = ['trade_date', 'ts_code'])
data.rename(columns = {'close':'far_price_ih'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data = pd.merge(data, price_if, left_on = ['trade_date', 'far_contract_if'], right_on = ['trade_date', 'ts_code'])
data.rename(columns = {'close':'far_price_if'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data = pd.merge(data, price_ic, left_on = ['trade_date', 'far_contract_ic'], right_on = ['trade_date', 'ts_code'])
data.rename(columns = {'close':'far_price_ic'},inplace=True)
data.drop(columns='ts_code',inplace = True)

data['near_maturity'] = np.array(map(date_diff, data['trade_date'], data['near_delist_date']))
data['far_maturity'] = np.array(map(date_diff, data['trade_date'], data['far_delist_date']))
data['near_maturity'] = data['near_maturity'].astype(int)
data['far_maturity'] = data['far_maturity'].astype(int)

data
```

```python
domestic_ir = pro.shibor(**{}, fields=["date","on"])
us_ir = pro.us_tycr(**{}, fields=["date","m1","y1","y10"])

data = pd.merge(data, domestic_ir, left_on = 'trade_date', right_on = 'date')
data.drop(columns=['date'], inplace=True)

data = pd.merge(data, us_ir, left_on = 'trade_date', right_on = 'date')
data.drop(columns=['date'], inplace=True)
```
